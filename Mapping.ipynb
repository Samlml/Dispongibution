```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
import xarray as xr
import cartopy.crs as ccrs
from maptools import make_map
```

```python
df=pd.read_csv('deep_sea_coral_sponge_data.csv')
```

```python
demo_df=df[df.VernacularNameCategory=='demosponge']
glass_df=df[df.VernacularNameCategory=='glass sponge']
```

```python
plt.figure(figsize=(10*1.65,10))
ax=make_map(projection=ccrs.Mercator())
ax.coastlines()
plt.plot(df.longitude,df.latitude,'.',color='r',transform=ccrs.PlateCarree())
plt.plot(glass_df.longitude,glass_df.latitude,'.',color='c',label='Glass Sponge',transform=ccrs.PlateCarree())
plt.plot(demo_df.longitude,demo_df.latitude,'.',color='y',label='Demosponge',transform=ccrs.PlateCarree())
plt.plot(glass_df.longitude,glass_df.latitude,'.',color='c',alpha=.3,label='Glass Sponge',transform=ccrs.PlateCarree())
```
